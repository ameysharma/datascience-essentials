**Neural Networks**

Neural networks are foundational to deep learning and machine learning, enabling computers to learn patterns and make intelligent decisions. Below is a detailed breakdown of how they work, their components, and how they're applied.

**1.1 Overview of Neural Networks**

**Input Layer:** The layer that receives the initial data.

**Hidden Layers:** Layers between the input and output layers where computations are performed. There can be multiple hidden layers.

**Output Layer:** The layer that produces the final output.




```python
# Neural Networks Explainer

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.colors import ListedColormap
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from IPython.display import display, Markdown

# Function to display theoretical explanations
def display_theory():
    display(Markdown("""
    ## Neural Networks Explained

    Neural networks are machine learning models inspired by the human brain. They consist of layers of interconnected nodes (neurons) that process data and learn patterns.

    ### Key Components
    - **Input Layer**: Takes input features.
    - **Hidden Layers**: Perform computations using weights, biases, and activation functions.
    - **Output Layer**: Produces final predictions.

    ### How Neural Networks Work
    1. **Forward Propagation**: Data flows through layers.
    2. **Loss Calculation**: Compares predicted and actual values.
    3. **Backward Propagation**: Adjusts weights and biases to minimize loss.

    ### Activation Functions
    - **ReLU**: \( f(x) = \max(0, x) \)
    - **Sigmoid**: \( f(x) = \frac{1}{1 + e^{-x}} \)
    - **Tanh**: \( f(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} \)

    Let's visualize how a simple neural network classifies data.
    """))

# Function to generate synthetic data
def generate_data():
    X, y = make_moons(n_samples=100, noise=0.2, random_state=42)
    return train_test_split(X, y, test_size=0.3, random_state=42)

# Function to visualize data
def visualize_data(X_train, y_train):
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='viridis')
    plt.title('Training Data')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()

# Function to train a simple neural network
def train_nn(X_train, y_train):
    clf = MLPClassifier(hidden_layer_sizes=(10,), max_iter=2000, random_state=42)
    clf.fit(X_train, y_train)
    return clf

# Function to visualize decision boundaries
def plot_decision_boundary(clf, X, y):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                         np.arange(y_min, y_max, 0.1))
    
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plt.contourf(xx, yy, Z, alpha=0.8, cmap=ListedColormap(['#FFAAAA', '#AAAAFF']))
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap='viridis')
    plt.title('Decision Boundary')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()

# Function to animate training process
def animate_training(clf, X_train, y_train):
    fig, ax = plt.subplots()
    
    def update(i):
        clf.partial_fit(X_train, y_train, classes=np.unique(y_train))
        ax.clear()
        plot_decision_boundary(clf, X_train, y_train)

    ani = animation.FuncAnimation(fig, update, frames=10, repeat=False)
    plt.show()

# Main function to run the explainer
def main():
    display_theory()
    
    # Generate data
    X_train, X_test, y_train, y_test = generate_data()
    visualize_data(X_train, y_train)

    # Train neural network
    clf = train_nn(X_train, y_train)

    # Plot decision boundary
    plot_decision_boundary(clf, np.vstack((X_train, X_test)), np.hstack((y_train, y_test)))

    # Animate training process
    animate_training(clf, X_train, y_train)

# Run the notebook main function
if __name__ == "__main__":
    main()
```

```python
# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Plot activation functions
x = np.linspace(-10, 10, 400)
fig, ax = plt.subplots(1, 3, figsize=(18, 5))

# Sigmoid
ax[0].plot(x, sigmoid(x))
ax[0].set_title("Sigmoid Activation Function")
ax[0].grid(True)

# ReLU
ax[1].plot(x, relu(x))
ax[1].set_title("ReLU Activation Function")
ax[1].grid(True)

# Tanh
ax[2].plot(x, tanh(x))
ax[2].set_title("Tanh Activation Function")
ax[2].grid(True)

plt.show()
```
